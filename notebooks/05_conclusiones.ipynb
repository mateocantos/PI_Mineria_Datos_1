## Objetivos y preguntas del proyecto

El proyecto buscó comprender el comportamiento de los usuarios de la plataforma de streaming. Las preguntas que guiaron el análisis fueron:

1. ¿Cómo se distribuye el tiempo mensual de visualización?
2. ¿Cómo se reparten los usuarios entre los planes de suscripción?
3. ¿La edad se relaciona con el tiempo de visualización?
4. ¿El tiempo de visualización varía según el plan?
5. ¿Varía según el género favorito?
6. ¿Cómo se relacionan entre sí las variables numéricas?

## Conclusiones por pregunta

### 1. Distribución del tiempo de visualización

- **Evidencia:** la distribución es asimétrica hacia la derecha; la mayoría de los usuarios consume entre ~500 y ~1200 minutos al mes (media 801, mediana 772), con una minoría de consumos muy altos que se conservaron por ser válidos.
- **Interpretación:** el consumo típico es moderado, pero un grupo reducido de usuarios muy intensivos estira el promedio por encima de la mediana.
- **Conclusión:** el tiempo de visualización se concentra en valores medios, con una cola de usuarios de alto consumo.

### 2. Distribución de los planes de suscripción

- **Evidencia:** Básico 44,9%, Estándar 35,3% y Premium 19,8%.
- **Interpretación:** predominan los planes de menor costo; Premium es minoritario aunque representa una porción relevante de usuarios.
- **Conclusión:** la base de usuarios se concentra en los planes más económicos.

### 3. Edad y tiempo de visualización

- **Evidencia:** el coeficiente de correlación de Pearson fue 0,005 y la recta de regresión resultó prácticamente horizontal.
- **Interpretación:** la edad no aporta información para explicar cuánto consume un usuario.
- **Conclusión:** la edad **no es un factor determinante** del tiempo de visualización.

### 4. Tiempo de visualización según el plan

- **Evidencia:** el consumo promedio crece con el nivel del plan: Básico 597, Estándar 872 y Premium 1139 minutos; la misma tendencia se observa en las medianas y los cuartiles.
- **Interpretación:** a mayor nivel de plan, mayor consumo, aunque las distribuciones se superponen, por lo que el plan no es el único factor que influye.
- **Conclusión:** el plan de suscripción **sí se asocia** con el tiempo de visualización; es la relación más clara de todo el análisis.

### 5. Tiempo de visualización según el género favorito

- **Evidencia:** las distribuciones de tiempo son muy similares entre los distintos géneros, con medianas comparables y amplia superposición.
- **Interpretación:** el género favorito no diferencia el nivel de consumo de los usuarios.
- **Conclusión:** el género favorito **no incide** en el tiempo de visualización.

### 6. Relaciones entre las variables numéricas

- **Evidencia:** la matriz de correlación mostró valores cercanos a cero entre edad, tiempo de visualización y tickets de soporte.
- **Interpretación:** las variables numéricas son prácticamente independientes entre sí.
- **Conclusión:** ninguna variable numérica explica a otra; el comportamiento de consumo depende de múltiples factores y no de una sola variable.

## Aporte del PCA

- **Evidencia:** la varianza explicada se repartió de forma pareja entre las tres componentes (aproximadamente 34%, 33% y 33%). En las cargas, PC1 se asocia sobre todo a la edad y los tickets, PC2 al tiempo de visualización y PC3 combina las tres.
- **Interpretación:** como las variables casi no se correlacionan, cada una aporta una dimensión propia y no hay redundancia que comprimir.
- **Conclusión:** el PCA **confirma lo observado en el EDA** (variables independientes) y muestra que, con estas tres variables, no se logra una reducción de dimensionalidad efectiva.

## Limitaciones

Aspectos del dataset, del proceso o del alcance que **restringen** lo que puede concluirse (no son hallazgos):

- El análisis es **descriptivo y exploratorio**: no se aplicaron pruebas estadísticas de significancia ni modelos, por lo que las asociaciones observadas (por ejemplo, plan–tiempo) son descriptivas y no causales.
- El alcance de las conclusiones está condicionado por la **información disponible** y por las decisiones documentadas durante la limpieza (imputaciones por mediana, conservación de fechas como NaT).
- **486 fechas** de último ingreso quedaron sin valor (NaT), lo que limita cualquier análisis de tipo temporal.
- El **PCA** trabaja con solo tres variables numéricas y sin correlación entre ellas, por lo que su capacidad de reducción es naturalmente limitada.
